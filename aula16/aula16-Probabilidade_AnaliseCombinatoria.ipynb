## Análise combinatória

```python
def perm(n): # calcula o número de permutações do conjunto n
    result = 1;
    if (n > 0):
        for i in range(n):
            result *= i+1
    return result

def pperm(n, k): # calcula o número de permutações parciais do conjunto n de tamanho k
    return (perm(n)/perm(n-k))
```
